```python
from sklearn import tree#module with packages, for normal ppl to use easily, decision rclassifier 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pet = pd.read_csv('petdata.csv')
```

```python
pet
```

```python
from sklearn.model_selection import train_test_split
X=pet.drop(['name', 'description','popularity','temperament','group','grooming_frequency_category','shedding_category','energy_level_category','trainability_category','demeanor_category'],axis=1)
y=pet['name']

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
dt = tree.DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_train)
```

```python
y_pred
```

```python
plt.figure(figsize = (250,200))
tree.plot_tree(dt, filled = True, fontsize = 12)
```
